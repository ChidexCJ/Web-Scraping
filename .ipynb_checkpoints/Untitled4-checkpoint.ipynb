```python
import requests
from bs4 import BeautifulSoup
```

```python
baseurl = 'https://www.jumia.com.ng/'
```

```python
for page in range(1,51):    
    response = requests.get(f'https://www.jumia.com.ng/catalog/?q=samsung+phones&page={page}#catalog-listing')
    soup =BeautifulSoup(response.text, 'html.parser')
    items = soup.find_all('a', class_ = 'core')
    for i in items:
        for desc in i.find_all('h3', class_ ='name'):
            description.append(desc)
        for pricee.text in i.find_all('div', class_ = 'prc'):
            price.append(pricee)
```

```python
for page in range(1,51):    
    response = requests.get(f'https://www.jumia.com.ng/catalog/?q=samsung+phones&page={page}#catalog-listing')
    soup =BeautifulSoup(response.text, 'html.parser')
    for i in soup:
        descr = [i.text for i in soup.findAll('h3', class_ = 'name')] if [i.text for i in soup.findAll('h3', class_ = 'name')] else " "
        description.append(descr)
        pric = [i.text for i in soup.findAll('div', attrs = {'class':'prc'})] if [i.text for i in soup.findAll('div', attrs = {'class':'prc'})] else " "
        price.append(pric)
    page+=1
```
